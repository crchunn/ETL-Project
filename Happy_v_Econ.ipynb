```python
economic_csv = "Resources/datasets_127094_304788_economic_freedom_index2019_data.csv"
econ_df = pd.read_csv(economic_csv, encoding='ISO-8859-1')
econ_df.head()
```

```python
econ_cols = ["Country Name", "2019 Score", "Property Rights", "Judical Effectiveness", "Government Integrity", 
             "Population (Millions)", "GDP Growth Rate (%)", "GDP per Capita (PPP)", 
             "Unemployment (%)", "Inflation (%)", "Public Debt (% of GDP)"]
econ_transformed_df = econ_df[econ_cols].copy()
econ_transformed_df.rename(columns = {'Country Name': 'Country_Name', '2019 Score': '2019_Score', 
                                      'Property Rights': 'Property_Rights', 'Judical Effectiveness': 'Judicial_Effectiveness', 'Government Integrity': 'Government_Integrity',
                         'Population (Millions)': 'Population_Millions', 'GDP Growth Rate (%)': 'GDP_Growth_Rate_Percent',
                                     'GDP per Capita (PPP)': 'GDP_per_Capita_PPP', 
                                      'Unemployment (%)': 'Unemployment_Percent', 'Inflation (%)': 'Inflation_Percent',
                                     'Public Debt (% of GDP)': 'Public_Debt_Percent_of_GDP'}, inplace = True)


econ_transformed_df.head()
```
